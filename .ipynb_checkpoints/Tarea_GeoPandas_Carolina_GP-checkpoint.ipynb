## **2. Definir variables:**

```python
# Cargar el archivo de datos 
archivo_datos = "datos/ara-ambiguus-cr.csv"

# Asignar a una variable el nombre que se desplegará en los gráficos estadísticos
nombre_cientifico = "Ara Ambiguus"

# Cantidad máxima de registros que se despliegan en un GeoDataFrame
pd.options.display.max_rows = 12
```

```python
# Selección de un subconjunto de columnas, se eligen las mismas al ejemplo de la clase
registros_presencia_df = registros_presencia_df[['species', 'stateProvince', 'locality', 'decimalLongitude', 'decimalLatitude', 'eventDate']]
registros_presencia_df
```

## 4. Análisis por medio de gráficos de la información:

#### 4.1 Graficación de los datos:

```python
# Estilo de los gráficos

plt.style.use('ggplot')
```

#### 4.2 Distribución de registros por mes:

```python
# Agrupación de los registros por mes
registros_x_mes = registros_presencia_df.groupby(registros_presencia_df['eventDate'].dt.month).count().eventDate

registros_x_mes
```

```python
# Reemplazo del número del mes por el nombre del mes
registros_x_mes.index=[calendar.month_name[x] for x in range(1,13)]

registros_x_mes
```

```python
# Gráfico de barras
registros_x_mes.plot(kind='bar',figsize=(13,8), color='purple', alpha=0.5)

# Título y leyendas en los ejes
plt.title('Registros de presencia de ' + nombre_cientifico + ' en Costa Rica en el año 2018', fontsize=22)
plt.xlabel('Mes', fontsize=18)
plt.ylabel('Cantidad de registros', fontsize=18);
```

#### 4.3 Distribución de registros por año:

```python
# Agrupación de los registros por año

registros_x_año = registros_presencia_df.groupby(registros_presencia_df['eventDate'].dt.year).count().eventDate

registros_x_año
```

```python
# Gráfico de barras
registros_x_año.plot(kind='bar',figsize=(13,8), color='blue', alpha=0.5)

# Título y leyendas en los ejes
plt.title('Registros de presencia de ' + nombre_cientifico + ' en Costa Rica según año de estudio', fontsize=22)
plt.xlabel('Año', fontsize=18)
plt.ylabel('Cantidad de registros', fontsize=18);
```

## 5. Mapeo de la información:

#### 5.1 GeoPandas:

```python
# Geodataframe
registros_presencia_geom = [Point(xy) for xy in zip(registros_presencia_df["decimalLongitude"], registros_presencia_df["decimalLatitude"])]
registros_presencia_gdf=gpd.GeoDataFrame(registros_presencia_df, crs={"init": "epsg:4326"}, geometry=registros_presencia_geom)
```

```python
registros_presencia_df
```

#### 5.2 Mapas con Folium:

```python
# Mapa centrado en Costa Rica
registros_presencia_mapa = folium.Map(location=[9.63, -84], zoom_start=8)

registros_presencia_mapa
```

##### 5.3 Mapa de distriución Ara ambiguus:

```python
# Mapa centrado en Costa Rica
registros_presencia_mapa = folium.Map(location=[10, -84], width=900, height=550, zoom_start=7, control_scale=True)

#Añadir capas base al mapa

folium.TileLayer(
    tiles='CartoDB positron', 
    name='CartoDB positron').add_to(registros_presencia_mapa)

# ESRI World Imagery
folium.TileLayer(
    tiles='http://services.arcgisonline.com/arcgis/rest/services/World_Imagery/MapServer/MapServer/tile/{z}/{y}/{x}',
    name='ESRI World Imagery',
    attr='ESRI World Imagery').add_to(registros_presencia_mapa)

# Añade al mapa mediante el método add_to()
folium.LayerControl().add_to(registros_presencia_mapa)

# Mapeo de los puntos correspondientes a los registros de presencia
for lat, lng, label in zip(registros_presencia_df.decimalLatitude, registros_presencia_df.decimalLongitude, registros_presencia_df.eventDate.astype(str)):
    folium.CircleMarker(
        [lat, lng],
        radius=3,
        color='green',
        fill=True,
        popup=label,
        fill_color='darkred',
        fill_opacity=0.6
    ).add_to(registros_presencia_mapa)
    
registros_presencia_mapa
```

##### 5.4 Mapa de "cluster" de Ara ambiguus:

```python
# Mapa centrado en Costa Rica
registros_presencia_mapa = folium.Map(location=[10, -84], width=900, height=550, zoom_start=7, control_scale=True)

# Instantiate a mark cluster object for the car accidents
registros_presencia_cluster = plugins.MarkerCluster(name='Cluster').add_to(registros_presencia_mapa)

for lat, lng, label in zip(registros_presencia_df.decimalLatitude, registros_presencia_df.decimalLongitude, registros_presencia_df.eventDate.astype(str)):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
        ).add_to(registros_presencia_cluster)
    
    
#Añadir capas base al mapa

folium.TileLayer(
    tiles='CartoDB positron', 
    name='CartoDB positron').add_to(registros_presencia_mapa)

# ESRI World Imagery
folium.TileLayer(
    tiles='http://services.arcgisonline.com/arcgis/rest/services/World_Imagery/MapServer/MapServer/tile/{z}/{y}/{x}',
    name='ESRI World Imagery',
    attr='ESRI World Imagery').add_to(registros_presencia_mapa)

# Añade al mapa mediante el método add_to()
folium.LayerControl().add_to(registros_presencia_mapa)
    
registros_presencia_mapa
```
